# L09/dataanalyse

## Qa Beskrivelse af datasæt til O4 projekt
For the final cpurse project we have for now choosen to investigate the pattern connecting the usage of smartphones and their battery drainage. We have found a dataset which should fit nicely with a variety of features to consider as well as a suitable amount of samples. We have found the dataset on [this kaggle.com website](https://www.kaggle.com/datasets/shayanzk/smartphone-battery-drain-and-usage-behavior/data?fbclid=IwY2xjawQaTFRleHRuA2FlbQIxMQBzcnRjBmFwcF9pZAEwAAEenuADs8Zv0yXkETBVfflt2mE4Z2rxNu7G04WoT-IRevJ2GDxlU1vKNgI2R0M_aem_ks0IsOAbeX272oRVYxVtvA). It is a common frustation and even with newer phones there seems to be so many factors at play that a pattern is not always descernable in the moment, thus we aim to machinelearn how these factors control the draining of smartphone batteries. The chosen dataset contains 10 different features covering, App, Screen on time, CPU usage in percent, Battery temperature, Battery drop pr Hour, Network type, Brightness level in percent, Ram usage in MB, whether the phone is charging or not, and lastly a usage mode which is in catagories of Music, Socail Media, Navigation, Video streaming, and Browsing.
The goal is to predict the battery drainage using machinelearning given some values of the above mentioned.
